This is a programe based on linear regression to predict movie revenue.  It is trained on the data obtained from a website called the-numbers.com and is stored in a .csv formate

```python

import pandas                         #pandas is a module=manipulating and analyzing data

from pandas import DataFrame          #dataframe is more like a powerful spreadsheet+part of the pandas module

import matplotlib.pyplot as plt       #matplotlib is a module(its a plotting library to graphs and charts)

from sklearn.linear_model import LinearRegression    # using a module called sckitlearn, used for classification+clusstering
```

```python

# we are telling the pandas module to read the csv file using ".read_csv". "data" is a an object to store the data in the .csv file

data = pandas.read_csv('cost_revenue_updated.csv')
```

```python
# it describes the data/object(.describe).  the numbers are in scientific notation

data.describe()  
```

```python

X = DataFrame(data, columns=['production_budget_usd'])     #X an object/datafraeme to hold production budget +we are seperating a specific column from the raw data.

y = DataFrame(data, columns=['worldwide_gross_usd'])      #y an object/dataframe to hold revenue+we are seperating a specific column from the raw data.
```

```python

plt.figure(figsize=(10,6))        #this will increase the size of the chart. we provide width and height

plt.scatter(X, y, alpha=0.3)               #this line of code will generate a scatte plot/graph in the output+(alpha) is for tranparency in the chart

plt.title('Film Cost vs Global Revenue')     #this is how title is added to the chart

plt.xlabel('Production Budget $')            #this is to add label to x axis

plt.ylabel('World wide Gross $')             #this is to add label to y axis

plt.ylim(0, 3000000000)                      #this is to remove the empty spaces for negetive numbers in the chart.we specify a limit for the yaxis

plt.xlim(0, 450000000)                       #this is to remove the empty spaces for negetive numbers in the chart.we specify a limit for the xaxis

plt.show()                      #this line will show the graph/chart of the data
```

```python

regression = LinearRegression()      #this line creates an object and use/refer that object as Linear Regression

regression.fit(X, y)                 #this line will crunch the numbers in the data. and find linear regression for us.
```

the regression coefficent show that for each dollar we spent we get three back.
so the relationship is somewhat plausable and trustworthy.

```python

#this line will find/print the slope of the line/coefficent of the regression
regression.coef_
```

according to the regression coefficient, if we make a movie with let say 1k then we will loose 7M 
it does not make any sense
this is becouse the model is trained on the data in which movies contains a very big budget
also it is calculated for all the training data

```python

#this line will find/print the intercept value of the line
regression.intercept_
```

```python

plt.figure(figsize=(10,6))        #this will increase the size of the chart. we provide width and height

plt.scatter(X, y, alpha=0.3)               #this line of code will generate a scatte plot/graph in the output+(alpha) is for tranparency in the chart

plt.plot(X, regression.predict(X), color='red' , linewidth=4)         #this will plot a regressional line on the data and also calculate the each value of x.
  # at the end of the above line, the color and width of the regressional line is changed for better visuall display

plt.title('Film Cost vs Global Revenue')     #this is how title is added to the chart

plt.xlabel('Production Budget $')            #this is to add label to x axis

plt.ylabel('World wide Gross $')             #this is to add label to y axis

plt.ylim(0, 3000000000)                      #this is to remove the empty spaces for negetive numbers in the chart.we specify a limit for the yaxis

plt.xlim(0, 450000000)                       #this is to remove the empty spaces for negetive numbers in the chart.we specify a limit for the xaxis

plt.show()                      #this line will show the graph/chart of the data
```

this single feature model explains 54% of the variation in the budget and gross revenue across the globe

```python

regression.score(X, y)    #this line will show how much the movie budget is related to movie revenue.
                          #this also shows the %age of correct prediction. in this case it is 54%
```
